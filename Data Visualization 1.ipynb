```python
import numpy as np
import pandas as pd

df = pd.read_csv('Topic_Survey_Assignment.csv')
```

```python
df
```

```python
#Sort the dataframe in descending order of Very interested
df.sort_values(by=['Very interested'], axis=0, ascending=False, inplace=True)
#df.rename(columns={'Unnamed: 0':'Course'}, inplace=True)
df.columns = list(map(str, df.columns))
df.set_index('Unnamed: 0', inplace=True)
df
```

```python
#Convert the numbers into percentages of the total number of respondents. 
#Recall that 2,233 respondents completed the survey. Round percentages to 2 decimal places
# df['Very Interested'] = df['Very interested']/2233
# df['Somewhat Interested'] = df['Somewhat interested']/2233
# df['Not Interested'] = df['Not interested']/2233
# df.drop(['Very interested', 'Somewhat interested', 'Not interested'], axis=1, inplace=True)
# df.round(2)
df = round((df/2233)*100,2)
df
```

```python
# use the inline backend to generate the plots within the browser
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot') # optional: for ggplot-like style

# check for latest version of Matplotlib
print ('Matplotlib version: ', mpl.__version__) # >= 2.0.0
```

```python
# plot chart
# use a figure size of (20, 8),
# bar width of 0.8,
# use color #5cb85c for the Very interested bars, color #5bc0de for the Somewhat interested bars, and color #d9534f for the Not interested bars,
# use font size 14 for the bar labels, percentages, and legend,
# use font size 16 for the title, and,
# display the percentages above the bars as shown above, and remove the left, top, and right borders.
ax = df.plot(kind='bar',
             figsize=(20,8),
             width=0.8,
             color=['#5cb85c','#5bc0de','#d9534f'],
             fontsize=14)

ax.set_title('Percentage of Respondents\' Interest in Data Science Areas', fontsize=16)

ax.set_facecolor('white')

ax.legend(fontsize=14,facecolor = 'white')

ax.get_yaxis().set_visible(False)

for p in ax.patches:
    ax.annotate(np.round(p.get_height(),decimals=2), 
                (p.get_x()+p.get_width()/2., p.get_height()), 
                ha='center', 
                va='center', 
                xytext=(0, 10), 
                textcoords='offset points',
                fontsize = 14
               )

ax.set_xlabel('',fontsize=14)

# plt.legend(labels=df.columns, fontsize=14)
# plt.title('Percentage of Respondents\' Interest in Data Science Areas', fontsize=16)
# plt.xticks(fontsize=14)
# plt.yticks([])

# plt.xlabel('',fontsize=14)
# plt.ylabel('',fontsize=14)

plt.show()
```
